```python
class Solution(object):
    def findNthDigit(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n < 10:
            return n
        
        base = 1
        x = 1
        while True:
            block_sum = 9*base*x
            if n < block_sum:
                break
            n = n - block_sum
            base = base * 10
            x = x + 1
        
        index = n // x + base
        rest = n % x
        if rest == 0:
            return int(str(index-1)[-1])
        else:
            return int(str(index)[rest-1])
```

```python
s = Solution()
```

```python
for i in range(2890,3000):
    print(s.findNthDigit(i))
```
