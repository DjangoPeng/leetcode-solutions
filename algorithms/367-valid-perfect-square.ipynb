```python
class Solution(object):
    def isPerfectSquare(self, num):
        """
        :type num: int
        :rtype: bool
        """
        if num == 0:
            return False
        left = 1
        right = num
        while left <= right:
            mid = (left + right) // 2
            cur = mid**2
            if cur == num:
                return True
            elif cur > num:
                right = mid - 1
            else:
                left = mid + 1
        return False
```

```python
s = Solution()
```

```python
for i in range(100):
    print(i, s.isPerfectSquare(i))
```
