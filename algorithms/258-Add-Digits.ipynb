```python
class Solution(object):
    def addDigits(self, num):
        """
        :type num: int
        :rtype: int
        """
        if num == 0:
            return 0
        # Easy to find the rule
        return (num - 1) % 9 + 1
```

```python
def genNew(n):
    new_n = 0
    while n > 0:
        new_n = new_n + n % 10
        n = n // 10
    return new_n

for i in range(100, 200):
    n = i
    while n > 9:
        n = genNew(n)
    print(i, (i-1) % 9 + 1 == n)
```
